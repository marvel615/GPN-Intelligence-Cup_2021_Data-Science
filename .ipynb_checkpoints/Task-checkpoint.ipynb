```python
buys
```

```python
buys = buys.drop(['client_id', 'shop_id', 'date'], axis=1)
```

```python
buys = buys.set_index(buys['cheque_id'].values)
```

```python
buys = buys.sort_values(by='cheque_id')
```

```python
buys = buys.set_index(np.arange(len(buys)))
```

```python
buys
```

```python
'''Работаем с таблицей товаров (nomenclature)

Составляем матирцу 5301(len(nomenclature)) * m (товары, которые относятся к требуемому нам списку) :
сначала делаем матрицу из 0
пробегаемся по всем транзакциям (по чекам) и суммируем встречаемость одного товара с другим
'''
```

```python
nomenclature
```

```python
set(nomenclature['sku_group'])
```

```python
need = ['Вода', 'Сладкие Уранированные напитки, холодный чай', 'Кофейные напитки с молоком',
        'Энергетические напитки', 'Снеки', 'Соки и сокосодержащие напитки']
```

```python
need_nomenclature = nomenclature.loc[nomenclature['sku_group'].isin(need)]
```

```python
need_nomenclature
```

```python
set(need_nomenclature['sku_group']) #проверим, на всякий случай
```

```python
#из таблицы выше, нам нужны только товары, удовл. требуемым группам покупок
need_tovars = sorted(list(set(need_nomenclature['sku_id'])))
m = len(need_tovars) #мы же будем строить матрицу 5103*m
```

```python
print(need_tovars[:5])
print(need_tovars[-5:])
```

```python
all_tovars = sorted(list(set(nomenclature['sku_id'])))
#по сути это просто np.arange(len(nomenclature)), но будем соблюдать формальность
```

```python
print(len(all_tovars))
print(m)
```

```python
Matrix_head = np.zeros((len(all_tovars), m))
```

```python
Matrix_head_df = pd.DataFrame(Matrix_head, index=all_tovars, columns=need_tovars)
```

```python
Matrix_head_df
```

```python
#пробегаемся по всем чекам в транзакциях
#каждый товар в чеке сверяем с остальными позициями в этом же чеке
#таким образом считаем количество покупок в парах
#не забываем занулить элементы с повторяющимися индексами 
```

```python
cheque_list = sorted(list(set(buys['cheque_id'])))
```

```python
print(cheque_list[:7])
print(cheque_list[-7:])
```

```python
for k in cheque_list:
    now_ts = buys[buys['cheque_id']==k]
    now_ts = now_ts.set_index(np.arange(len(now_ts)))
    
    for i in now_ts.index:
        for j in now_ts.index:
            
            if now_ts['sku_id'][j] in need_tovars:
                idx1 = now_ts['sku_id'][j]
                idx2 = now_ts['sku_id'][i]
                
                Matrix_head_df[idx1][idx2] = Matrix_head_df[idx1][idx2] + 1
                
    if (k%50000 == 0): #Отслеживаем протекание процесса, точнее кол-во выполненных итераций. UPD: цикл длился около 9.5 часа
        print(k)
```

```python
Matrix_head_df[1882][1882]
```

```python
#чтобы не предлагать уже купленные товары клиентам
for i in Matrix_head_df.columns:
    Matrix_head_df[i][i] = 0
```

```python
Matrix_head_par = Matrix_head_df.copy()
Matrix_head_par.columns = [str(st) for st in Matrix_head_par.columns]
```

```python
Matrix_head_par.to_parquet('Matrix_freq.parquet',engine='pyarrow', index=False) #Сохраним, чтобы если что не пересчитывать
```

```python
#Теперь мы должны получить матрицу top20. Предлагаю индексы - все товары, столбцы - числа 1..20, соотв. топу
K=20 #занесем в переменную, чтобы соблюдать переносимость (вдруг кто-то захочет top10 отобрать)
```

```python
Matrix_head_trans = Matrix_head_df.transpose()
```

```python
Matrix_head_trans
```

```python
mat_top20_zeros = np.zeros((len(all_tovars), K))
```

```python
mat_top20_nan = np.full_like(mat_top20_zeros, np.nan)
```

```python
Martix_top20 = pd.DataFrame(mat_top20_nan, index=all_tovars, columns=np.arange(1, K+1))
```

```python
Martix_top20
```

```python
Martix_top20.isna().sum().sum()/5103
```

```python
Matrix_head_trans
```

```python
for col in Matrix_head_trans.columns:
    lst = sorted(Matrix_head_trans[col], reverse=True) #сортируем столбец весов по убыванию (так как нам нужны топ-позиции)
    
    now_ser = Matrix_head_trans[col]
    
    spisok_top = [] #тут будут значения sku_id нужных товаров, которые чаще продаются с соотв. товаром (col)
    spisok_idx = []
    for s in lst[:K]:
        help_ser = now_ser[now_ser == s] #выбираем все товары с текущим весом по отношению к нынешнему товару (col)
        
        
        for idx in help_ser.index:
            spisok_top.append(idx)
            spisok_idx.append(s)
            
        spisok_top_df = pd.DataFrame(spisok_top, index=spisok_idx, columns=['Top']) #создаем таблицу нужных товаров
        spisok_top_df.drop_duplicates(inplace=True) #так как могут повторяться, убираем повторы
        spisok_top_final = spisok_top_df.iloc[:K]['Top']
        
        top_place = spisok_top_final.values
        
    for k_top in Martix_top20.columns:
        Martix_top20[k_top][col] = top_place[k_top-1] #заполняем итоговую матрицу Топов
```

```python
Martix_top20 = Martix_top20.astype('int64') #значения должны быть типа int, так как это индексы sku_id
```

```python
Martix_top20
```

```python
Martix_top20_par = Martix_top20.copy()
Martix_top20_par.columns = [str(st) for st in Martix_top20_par.columns]
```

```python
Martix_top20_par.to_parquet('Top_Matrix.parquet',engine='pyarrow', index=False) #Сохраняем
```

```python
#Приступаем к финальной части задания
transactions_for_submission = pd.read_parquet('transactions-for_submission.parquet')
```

```python
transactions_for_submission
```

```python
transactions_for_submission = transactions_for_submission.set_index(transactions_for_submission['cheque_id'].values)
```

```python
transactions_for_submission = transactions_for_submission.sort_values(by='cheque_id')
```

```python
transactions_for_submission
```

```python
idx_subm = np.unique(transactions_for_submission.index) #не забываем что у нас таблица выше упорядочена по чекам
```

```python
idx_subm
```

```python
len(idx_subm)
```

```python
len(idx_subm)*K #длина таблицы с ответом
```

```python
answer = pd.DataFrame(np.zeros((len(idx_subm)*K,3)), columns=['cheque_id', 'sku_id', 'position'])
```

```python
answer
```

```python
pos = np.zeros(len(idx_subm)*K)
```

```python
for ixs in range(len(pos)):
    pos[ixs] = (ixs % 20) + 1
```

```python
answer['position'] = pos
```

```python
answer = answer.astype('int64')
```

```python
answer.iloc[:22]
```

```python
#Рейтинг готов. Теперь заполняем столбцы cheque_id и sku_id
```

```python
counter = 0
p_start = 0

for idx in idx_subm:
    check_ts = transactions_for_submission.loc[idx] 
    need_idxes = np.array(check_ts['sku_id']) #товары в текущем чеке
    top_now = Martix_top20.iloc[need_idxes] #таблица с топом релевантных товаров для каждого товара в текущем чеке
    top_now = top_now.drop_duplicates()
    
    if type(top_now) == pd.core.series.Series:
        top_now = top_now.to_frame()
        top_now = top_now.transpose()
    
    w_cols = [] #это у нас столбцы для матрицы весов в текущем чеке
    for t in top_now.columns:
        for i in top_now.index:
            w_cols.append(top_now[t][i])
    
    
    w_cols = np.unique(w_cols) #чтобы не было повторений
    index_w = np.array(top_now.index) #а это строки для матрицы весов в текущем чеке
    zeros_w = np.zeros((len(index_w), len(w_cols)))
    
    #weights - будущая матрица весов для товаров в текущем чеке и их релевантными товарами
    weights = pd.DataFrame(zeros_w, index=index_w, columns=w_cols)
    
    #заполняем weights
    for col in weights.columns:
        for ix in weights.index:
            weights[col][ix] = Matrix_head_df[col][ix]
            
    #теперь надо просуммировать веса по колонкам и упорядочить их по убыванию.
    sum_weights = []
    for col in weights.columns:
        sum_weights.append(weights[col].sum())
        
    weights_for_cheque = pd.DataFrame(sum_weights, index=weights.columns, columns=['weights_cheque'])
    weights_for_cheque = weights_for_cheque.sort_values(by='weights_cheque', ascending = False) #лучшие - наверху
    weights_for_cheque = weights_for_cheque.iloc[:K] #нам же надо топ20 (K=20)
    
    #weights_for_cheque.index - то что нам надо. Это sku_id
    tovar_tops = np.array(weights_for_cheque.index)
    for top in range(K):
        answer['cheque_id'][p_start + top] = idx
        answer['sku_id'][p_start + top] = tovar_tops[top]
        
    p_start = p_start + K
    
    counter = counter + 1
    if (counter % 50000 == 0): #Отслеживаем работу алгоритма. Затраченное время - примерно 2.5 часа
        print(counter)
```
